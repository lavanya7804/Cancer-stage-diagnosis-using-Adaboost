# Import dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import the Dataset

```python
data=pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

# Preprocessing

```python
from sklearn import preprocessing
```

```python
e=preprocessing.LabelEncoder()
diagnosis_encoded=e.fit_transform(data['diagnosis'])
data['diagnosis']=diagnosis_encoded
```

```python
data.head()
```

# Assigning dependent and independent variable

```python
x = data.iloc[:, 2:-1]
```

```python
x.head()
```

```python
y=data.diagnosis 
```

```python
y
```

# Splitting the Dataset into Training and Testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

# fitting the Model

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
base_learner = DecisionTreeClassifier(max_depth=1)
adaboost = AdaBoostClassifier(base_estimator=base_learner, n_estimators=50, random_state=42)
```

```python
adaboost.fit(x_train, y_train)
```

```python
predictions = adaboost.predict(x_test)
predictions 
```

# Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

```python
plt.plot([accuracy], marker='o', linestyle='-', color='b')
plt.title('Accuracy Plot')
plt.xlabel('Iterations or Epochs')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```
